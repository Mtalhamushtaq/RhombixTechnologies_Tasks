```python

import numpy as np
import pandas as pd
from scipy import stats


np.random.seed(42)


clicks_A = np.random.randint(20, 60, 100)   
clicks_B = np.random.randint(70, 130, 100) 

sales_A = clicks_A // 5 + np.random.randint(0, 5, 100)  
sales_B = clicks_B // 4 + np.random.randint(0, 8, 100)   


result = stats.ttest_ind(clicks_A, clicks_B)


print("A/B TEST RESULT")
print("===============")
print(f"Campaign A clicks = {clicks_A.sum()} | Sales = {sales_A.sum()}")
print(f"Campaign B clicks = {clicks_B.sum()} | Sales = {sales_B.sum()}")
print()
print(f"P-value = {result.pvalue:.10f}")

if result.pvalue < 0.05:
    print("WINNER = CAMPAIGN B (100% sure!)")
    print("Tell your boss: STOP Campaign A, USE ONLY CAMPAIGN B!")
else:
    print("No winner")


print()
print("=== COPY THIS TO YOUR REPORT ===")
print("We tested Campaign A vs Campaign B with 200 users.")
print("Campaign B got way more clicks and sales.")
print("Statistical test (p-value < 0.05) proves Campaign B is better.")
print("Recommendation: Use Campaign B only.")
print("=================================")
```
