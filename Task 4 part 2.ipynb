```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\ESHOP\Downloads\sales_data.csv")
df.head(3)
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
```

```python
total_revenue = df['Revenue'].sum()
print("Total Revenue:", total_revenue)
```

```python
revenue_by_product = df.groupby('Product')['Revenue'].sum()
print(revenue_by_product)
```

```python
revenue_by_product.plot(kind='bar', title='Revenue by Product')
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.tight_layout()

plt.savefig("revenue_by_product.png")
plt.show()
```

```python
revenue_by_product.to_csv("revenue_summary.csv")

summary = pd.DataFrame({
    "Metric": ["Total Revenue"],
    "Value": [total_revenue]
})

summary.to_csv("sales_summary.csv", index=False)
```
